```python
import cv2

video_path = "/home/ducca/C2KD/ave/utils/data/AVE_Dataset/AVE/VWi2ENBuTbw.mp4"
vid = cv2.VideoCapture(video_path)
fps = int(vid.get(cv2.CAP_PROP_FPS))
video_frames = vid.get(cv2.CAP_PROP_FRAME_COUNT)
video_frames
```

```python
import os
import numpy as np
frame_kept_per_second = 1

def video2frame_update_SE( frame_save_path = "/home/ducca/C2KD/ave/utils/data/AVE_Dataset/Image-01-FPS-SE/VWi2ENBuTbw", min_save_frame=3, start_t=0, end_t=10):
        frame_save_path = frame_save_path #path

        count = 0
        save_count = 0
        frame_interval = int(fps / frame_kept_per_second) # 30

        num_count = 0
        while count < video_frames: #300
            ret, image = vid.read()
            if not ret:
                break
            if count % fps == 0:
                frame_id = 0
            if frame_id < frame_interval * frame_kept_per_second and frame_id % frame_interval == 0:
                if start_t <= num_count <= end_t:
                    # print('save')
                    save_name = '{0}/{1:05d}.jpg'.format(frame_save_path, count)
                    cv2.imencode('.jpg', image)[1].tofile(save_name)
                    save_count += 1
                num_count += 1

            frame_id += 1
            count += 1

        if save_count < min_save_frame:
            add_count = min_save_frame - save_count
            count = 0
            if video_frames < min_save_frame:
                while count < add_count:
                    frame_id = np.random.randint(0, min_save_frame)
                    save_name = '{0}/{1:05d}.jpg'.format(frame_save_path, frame_id)
                    if not os.path.exists(save_name):
                        vid.set(cv2.CAP_PROP_POS_FRAMES, 0)
                        ret, image = vid.read()
                        cv2.imencode('.jpg', image)[1].tofile(save_name)
                        count += 1
            else:
                while count < add_count:
                    frame_id = np.random.randint(start_t*fps, end_t*fps)
                    save_name = '{0}/{1:05d}.jpg'.format(frame_save_path, frame_id)
                    if not os.path.exists(save_name):
                        vid.set(cv2.CAP_PROP_POS_FRAMES, frame_id)
                        ret, image = vid.read()
                        cv2.imencode('.jpg', image)[1].tofile(save_name)
                        count += 1

video2frame_update_SE()
```
